VISUALIZING DATA

PIE PLOT OF SUB-CATEGORY

```python
plt.figure(figsize=(12,10))
store_df['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

The store has wide variety of office supplies especially in binders and paper departments

```python
store_df.groupby("City")["Sales"].sum().plot.line()
```

SALES PER SUB-CATEGORY

```python
#total profit and sales per sub-category
store_df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title("Total Profit and Sales per Sub-Category")
plt.show()
```

COUNT PLOT OF STATE

```python
plt.figure(figsize=(15,8))
sns.countplot(x=store_df['State'])
plt.xticks(rotation=90)
plt.show()
```

From the countplot, West Virginia has least count


LINE PLOT OF DISCOUNT VS PROFIT

```python
plt.figure(figsize=(10,4))
sns.lineplot(x='Discount',y='Profit',data=store_df,color='r',label='Discount')
plt.legend()
```

When discount is 0.5 the profit decreases

BAR PLOT OF SUBCATEGORY AND CATEGORY

```python
plt.figure(figsize=(16,8))
plt.bar( "Sub-Category","Category", data=store_df)
plt.show()
```

In furniture category the frequency of individual item is less
In office supplies category the frequency of individual item is less
In technology category the frequency of individual item is high


COUNT PLOT OF SEGMENT

```python
sns.countplot(x=store_df['Segment'])
```

TOP TEN BEST PERFORMING CITIES

```python
city_sales_df = pd.DataFrame(store_df.groupby(['City'])['Discount','Quantity','Profit'].mean().
                             sort_values('Discount',ascending = False))
top10 = city_sales_df[:10]
top10
```

```python
store_df.groupby("Ship Mode")["Sales"].sum().plot.line()
```

WHICH CUSTOMER SEGMENT IS THE MOST PROFITABLE?

```python
df4 = pd.DataFrame(store_df.groupby(['Segment'])[['Profit']].mean())
df4
```

```python
sns.barplot(data = df4, x = df4.index, y = df4.Profit, palette = "Blues")
plt.title("Customer Segment Profitability")
plt.show()
```

Home Office Segment is most profitable

```python
store_df['Quantity'].hist(bins=50,figsize=(15,8))
```

TOTAL PROFIT PER SEGMENT

```python
plt.figure(figsize=(12,3))
plt.barh(store_df['Segment'],store_df['Profit'],
        color = 'b',
        alpha = 1,
       height=0.5)

plt.title('Total Profit Per Segment',fontdict={'fontsize':15})
plt.ylabel('Profit')
plt.xlabel('Segment')
plt.show()

```

From the above graphs, it is realized that the consumer segment is doing positively well in terms of profit and discounted sales and the Home Office segment is the least performing in terms of discounted sales and profit
